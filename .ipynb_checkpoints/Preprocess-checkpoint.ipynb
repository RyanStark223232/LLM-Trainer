```python
data_list = list()
for i in range(len(df)-1):
    curr_row = df.loc[i, ['name', 'line']]
    next_row = df.loc[i+1, ['name', 'line']]
    print(next_row['name'])
    if next_row.['name'] == "Rick":
        row_dict = {
            "instruction": f"Answer the question from {curr_row.name} as Rick from Rick & Morty.",
            "input": f"{curr_row.line}",
            "output": f"{next_row.line}"
        }
        print(row_dict)
    break
```
